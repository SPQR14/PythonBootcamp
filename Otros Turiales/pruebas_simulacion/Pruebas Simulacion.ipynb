```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
from IPython.display import Math, Latex
from IPython.core.display import Image
sns.set(color_codes = True)
sns.set(rc={'figure.figsize':(18,10)})
import scipy as sc
```

```python
#Uniforme
n = 98280
inicio = 1
fin = 28
muestra_uniforme = sc.stats.uniform.rvs(size = n, loc = inicio, scale = fin)
```

```python
ax = sns.distplot(muestra_uniforme,
                 bins = 28,
                 kde = True,
                 color = 'skyblue',
                hist_kws={'linewidth': 15, 'alpha': 1})
ax.set(xlabel= "98,280 simulaciones", ylabel="Frecuencia")
```
